```python
%matplotlib inline
from kernlearn.learn import first_order_system, second_order_system
from kernlearn.utils import data_loader, train_test_split
from kernlearn.nn_models import *
```

```python
hparams = {
    "learning_rate": 1e-3,
    "n_epochs": 500,
    "optimizer": "adam",
    "layer_sizes": [8],
    "activation": "sigmoid",
}
seed = 0
```

```python
sheep_data1 = data_loader("data/sheep/json/processed_data1.json")
sheep_data2 = data_loader("data/sheep/json/processed_data2.json")
sheep_data3 = data_loader("data/sheep/json/processed_data3.json")
sheep_train1, sheep_test1 = train_test_split(sheep_data1, ind=28)
sheep_train2, sheep_test2 = train_test_split(sheep_data2, ind=28)
sheep_train3, sheep_test3 = train_test_split(sheep_data3, ind=28)
PP1NN3 = FirstOrderPredatorPreyNN3(
    hparams["layer_sizes"], hparams["activation"], seed=seed
)
out_PP1NN3_1 = first_order_system(
    PP1NN3, hparams, sheep_train1, sheep_test1, data_string="sheep/E1"
)
out_PP1NN3_2 = first_order_system(
    PP1NN3, hparams, sheep_train2, sheep_test2, data_string="sheep/E2"
)
out_PP1NN3_3 = first_order_system(
    PP1NN3, hparams, sheep_train3, sheep_test3, data_string="sheep/E3"
)
```
